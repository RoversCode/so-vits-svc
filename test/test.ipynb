```python
import torch
from eres2net.eres2net import ERes2NetV2
import torchaudio.compliance.kaldi as Kaldi


class FBank(object):
    def __init__(
        self,
        n_mels,
        sample_rate,
        mean_nor: bool = False,
    ):
        self.n_mels = n_mels
        self.sample_rate = sample_rate
        self.mean_nor = mean_nor

    def __call__(self, wav, dither=0):
        sr = 16000
        assert sr == self.sample_rate
        if len(wav.shape) == 1:
            wav = wav.unsqueeze(0)
        # select single channel
        if wav.shape[0] > 1:
            wav = wav[0, :]
        assert len(wav.shape) == 2 and wav.shape[0] == 1
        feat = Kaldi.fbank(
            wav, num_mel_bins=self.n_mels, sample_frequency=sr, dither=dither
        )
        # feat: [T, N]
        if self.mean_nor:
            feat = feat - feat.mean(0, keepdim=True)
        return feat
    

model_config = {
    "feat_dim": 80,
    "embedding_size": 192,
    "baseWidth": 26,
    "scale": 2,
    "expansion": 2,
}
model = ERes2NetV2(**model_config)

model.load_state_dict(
    torch.load(
        "/mnt/e/Workspace/growth/audio/so-vits-svc/preprocess/eres2net/eres2netv.pth"
    ),
    strict=True,
)
model = model.to("cuda")

feature_extractor = FBank(80, sample_rate=16000, mean_nor=True)
import torchaudio
def load_wav(wav_file, obj_fs=16000):
    wav, fs = torchaudio.load(wav_file)
    if fs != obj_fs:
        wav, fs = torchaudio.sox_effects.apply_effects_tensor(
            wav, fs, effects=[["rate", str(obj_fs)]]
        )
        if wav.shape[0] > 1:
            wav = wav[0, :].unsqueeze(0)
    return wav
```

```python
keys = list(keep.keys())
print(len(keys))
# values = keep[keys[9]]
# print(len(values))
import random
import numpy as np
import os

def cosine_similarity(arr1, arr2):
    # 确保输入的数组形状为[1, 192]

    assert arr1.shape == (1, 192), "arr1的形状应为[1, 192]"
    assert arr2.shape == (1, 192), "arr2的形状应为[1, 192]"

    # 计算数组的二范数(L2范数)
    norm1 = np.linalg.norm(arr1)
    norm2 = np.linalg.norm(arr2)

    # 计算数组的点积
    dot_product = np.dot(arr1, arr2.T)

    # 计算余弦相似度
    cosine_sim = dot_product / (norm1 * norm2)

    return cosine_sim[0][0]

audio_path = '/mnt/e/Workspace/growth/audio/so-vits-svc/preprocess/raw_data/蜡笔小新切片'
from tqdm import tqdm
results = {}
for i in tqdm(range(len(keys))):
    spk = keep[keys[i]]
    # 随机选择
    s = random.choice(spk)
    path = os.path.join(audio_path, s)
    try:
        s_wav = load_wav(path)
    except:
        continue
    input_a = feature_extractor(s_wav).unsqueeze(0)
    timbre_a = model(input_a.to('cuda')).detach().cpu().numpy()
    for j in range(i+1, len(keys)):
        spk_j = keep[keys[j]]
        for _ in range(2):
            t = random.choice(spk_j)
            path = os.path.join(audio_path, t)
            try:
                wav = load_wav(path)
            except:
                continue
            input_b = feature_extractor(wav).unsqueeze(0)
            timbre_b = model(input_b.to('cuda')).detach().cpu().numpy()
            if cosine_similarity(timbre_a, timbre_b) > 0.7:
                if keys[i] not in results:
                    results[keys[i]] = []
                results[keys[i]].append(keys[j])
                break
            

        
```
